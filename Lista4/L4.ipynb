```python
class Empty(Exception):
    pass
class LinkedStack:
    #--- -Node class- ---
    class _Node:
        __slots__ = '_element', '_next' #faster memory access

        def __init__(self,element,next):
            self._element = element
            self._next = next

    #--- -Stack methods- ---
    def __init__(self): #empty stack
        self._head = None
        self._size = 0

    def __len__(self):
        return self._size

    def is_empty(self):
        return self._size == 0

    def push(self,e):
        self._head = self._Node(e,self._head)
        self._size += 1

    def top(self):
        if self.is_empty():
            raise Empty('Stack is empty!')
        return self._head._element

    def pop(self):
        if self.is_empty():
            raise Empty('Stack is empty!')
        answer = self._head._element
        self._head = self._head._next
        self._size -= 1
        return answer
```

```python
class BinaryTree:
    def __init__(self, size):
        self.tree = [None] * size

    def set_root(self, value):
        if self.tree[0] is None:
            self.tree[0] = value
        else:
            print("Root already exists.")

    def add_left_child(self, parent_index, value):
        left_child_index = 2 * parent_index + 1

        if (
            parent_index < 0
            or parent_index >= len(self.tree)
            or self.tree[parent_index] is None
            or self.tree[left_child_index] is not None
        ):
            print("Invalid left child addition.")
        else:
            self.tree[left_child_index] = value

    def get_left_child(self, parent_index):
        left_child_index = 2 * parent_index + 1

        if (
            parent_index < 0
            or left_child_index >= len(self.tree)
            or self.tree[parent_index] is None
        ):
            return None
        else:
            return self.tree[left_child_index]

    def add_right_child(self, parent_index, value):
        right_child_index = 2 * parent_index + 2

        if (
            parent_index < 0
            or parent_index >= len(self.tree)
            or self.tree[parent_index] is None
            or self.tree[right_child_index] is not None
        ):
            print("Invalid right child addition.")
        else:
            self.tree[right_child_index] = value

    def get_right_child(self, parent_index):
        right_child_index = 2 * parent_index + 2

        if (
            parent_index < 0
            or right_child_index >= len(self.tree)
            or self.tree[parent_index] is None
        ):
            return None
        else:
            return self.tree[right_child_index]

    def remove_node(self, index):
        if index < 0 or index >= len(self.tree) or self.tree[index] is None:
            print("Invalid node removal.")
        else:
            self.tree[index] = None

    def __str__(self):
        return str(self.tree)
```

```python
class TreeNode:
    def __init__(self, value, left_child=None, right_child=None,root = None):
        self.value = value              # string, stored value
        self.left_child = left_child    # None or TreeNode instance
        self.right_child = right_child  # None or TreeNode instance

    def get_value(self):
        return self.value

    def add_left_child(self, value):
        if self.left_child is None:
            raise ValueError("Left Child already present")
        self.left_child = TreeNode(value)

    def add_right_child(self, value):
        if self.right_child is None:
            raise ValueError("Left Child already present")
        self.right_child = TreeNode(value)

    def get_left_child(self):
        return self.left_child

    def get_right_child(self):
        return self.right_child

    def has_left_child(self):
        if self.left_child is None:
            return False
        else:
            return True

    def has_right_child(self):
        if self.right_child is None:
            return False
        else:
            return True

    def is_leaf(self):
        return not(self.has_left_child() and self.has_right_child())

    def __str__(self):
        '''Represents node and its children as string.'''
        s = ''
        if self.is_leaf():
            return str(self.value)
        else:
            if self.value not in ['sin', 'cos', 'exp', 'log']:
                #Ponieważ te funkcje są reprezentowane w taki a nie inny sposób (rysunek), 
                #to żeby uniknąć wypisywania ich dwa razy wypisujemy tylko lewe dziecko
                s += str(self.value)
            if self.value in ['exp', 'log']: 
                #zmiana wymagana bo exp(x) i log(x) pokazywaly sie jako expx i logx
                s += '('
            if self.has_left_child():
                s = '(' + str(self.left_child) + s
            if self.has_right_child():
                s = s + str(self.right_child) + ')'
            if self.value in ['exp', 'log']:
                #zamykanie ) exp(x i log(x
                s += ')'
        return s

def mathToken(raw):
    """
    Funkcja tokenizująca wyrażenie matematyczne
    """
    SYMBOLS = set('^+-*/() ')
    FUNCTIONS = ['sin','cos','exp','log'] # allow for only '*' for multiplication as 'x' will be a variable
    mark = 0
    tokens = []
    n = len(raw)
    for j in range(n):
        if raw[j] in SYMBOLS:
            if mark != j:
                tokens.append(raw[mark:j]) # complete preceding token
            if raw[j] != ' ':
                tokens.append(raw[j]) # include this token
            mark = j+1 # update mark to being at next index
    if mark != n:
        tokens.append(raw[mark:n]) # complete preceding token
    #dodawanie funkcji sin, cos, exp i log
    for k in range(len(tokens)-3):
        if tokens[k] in FUNCTIONS and tokens[k+3]==')':
            tokens.insert(k+3,'+')
            tokens.insert(k+4,'0')

    return tokens


def build_expression_tree(tokens):
    #--- Code from lecture with added lines. ---
    '''Builds binary expression tree from given tokens.'''
    S = [] # we use Python list as stack
    for t in tokens:
        if t in '+-*/^': # t is an operator symbol
            S.append(t) # push the operator symbol
        elif t not in '()': # consider t to be a literal
            if t.isnumeric(): # ADDED to handle sin,cos,exp,log
                S.append(TreeNode(int(t))) # push trivial tree storing value
            else:
                S.append(TreeNode(t))
            if t in 'sin cos exp log' and len(t) == 3:
                S.append(t)
        elif t == ')': # compose a new tree from three constituent parts
            right = S.pop() # right subtree as per LIFO
            op = S.pop() # operator symbol
            left = S.pop() # left subtree
            S.append(TreeNode(op, left, right)) # repush tree
    # we ignore a left parenthesis
    return S.pop()

def deriveTree(tree, var):
    '''
    Funkcja rekurencyjnie obliczająca pochodną drzewa wyrażenia matematycznego, milion caseów dla każdej operacji matematycznej
    '''
    new_tree = TreeNode(None)
    if tree.is_leaf(): #basecase, kazda pochodna sprowadza się do pochodnej stałej lub pochodnej zmiennej
        if tree.value == var:
            new_tree = TreeNode(1) #pochodna zmiennej
        else:
            new_tree = TreeNode(0) #pochodna stałej
    else: #poniewaz budujemy drzewo w taki a nie inny sposob, jezeli nie jestesmy w lisciu to wartosc node'a na pewno jest operatorem
        if tree.value in '+-':
            #pochodna sumy i roznicy, raczej zrozumiale, pochodna sumy/roznicy to suma/roznica pochodnych
            new_tree.left_child = deriveTree(tree.left_child,var)
            new_tree.right_child = deriveTree(tree.right_child,var)
            new_tree.value = tree.value
        elif tree.value == '*':
            #pochodna iloczynu, wzor znany z analizy (a*b)' = a'b + ab'
            new_tree.left_child = TreeNode('*', deriveTree(tree.left_child,var), tree.right_child)
            new_tree.right_child = TreeNode('*', tree.left_child, deriveTree(tree.right_child,var))
            new_tree.value = '+'
        elif tree.value == "/":
            #pochodna ilorazu dana wzorem (a/b)' = (a'b-ab')/b^2, oczywiscie rozbite na drzewo (zob. rysunek)
            new_tree.left_child = TreeNode('-', TreeNode('*', deriveTree(tree.left_child,var), tree.right_child),TreeNode('*', tree.left_child, deriveTree(tree.right_child,var)))
            new_tree.right_child = TreeNode('^', tree.right_child, TreeNode(2))
            new_tree.value = '/'
        elif tree.value == "^":
            #pochodna potęgi, przy założeniu że nie jest to funkcja wykładnicza. Korzysta z reguły łańcuchowej.
            #(a^b)' = (b*(a^(b-1)))*a' - rozrysowane na drzewo w paincie
            new_tree.left_child = TreeNode('*', tree.right_child, TreeNode('^', tree.left_child, TreeNode('-', tree.right_child, TreeNode(1))))
            new_tree.right_child = deriveTree(tree.left_child,var)
            new_tree.value = '*'
        elif tree.value == 'sin':
            #pochodna sinusa, ktorego reprezentacja w drzewie jest nie do konca oczywista (podobnie z innymi funcjami, patrz rysunek)
            new_tree.left_child = TreeNode('cos', TreeNode('cos'), tree.right_child)
            new_tree.right_child = deriveTree(tree.right_child, var)
            new_tree.value = '*'
        elif tree.value == 'cos':
            #pochodna cosinusa, analogiczna do sinusa z dodanym - (0 - wyrazenie) z przodu.
            new_tree.left_child = TreeNode(0)
            new_tree.right_child = TreeNode('*', TreeNode('sin', TreeNode('sin'), tree.right_child), deriveTree(tree.right_child, var))
            new_tree.value = '-'
        elif tree.value == 'exp':
            #sprawdzamy czy mamy wyrazenie exp(log(x)), gdzie mozemy pominac liczenie pochodnej (zwraca x)
            if tree.right_child.value == 'log':
                if type(tree.right_child.right_child) is TreeNode:
                    new_tree = tree.right_child.right_child
                else:
                    new_tree = TreeNode(tree.right_child.right_child)
            else:
                #analogiczne do sin(x)
                new_tree.left_child = TreeNode('exp', TreeNode('exp'),tree.right_child)
                new_tree.right_child = deriveTree(tree.right_child,var)
                new_tree.value = '*'
        elif tree.value == 'log':
            #sprawdzamy czy mamy wyrazenie log(exp(x)), gdzie mozemy pominac liczenie pochodnej (zwraca x)
            if tree.right_child.value == 'exp':
                if type(tree.right_child.right_child) is TreeNode:
                    new_tree = tree.right_child.right_child
                else:
                    new_tree = TreeNode(tree.right_child.right_child)
            else:
                #pochodna logarytmu, log(x)' = (1/x)*x', gdzie x jest funkcją złożoną o zmiennej var
                new_tree.left_child = TreeNode('/', TreeNode(1), tree.right_child)
                new_tree.right_child = deriveTree(tree.right_child, var)
                new_tree.value = '*'
    return new_tree
    
#expr = "(((2*3)+(1*4))/(2+0))"
#expr = '((2*x)+(x^2))'
#expr = '(log(exp(x)))'
expr = '((x*9) + (sin(1)))'
tokeny = mathToken(expr)
drzewo = build_expression_tree(tokeny)
print(drzewo)
pochodna = deriveTree(drzewo, 'x')
print(pochodna)
```
