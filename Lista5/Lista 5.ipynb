Zad. 1 Jaki będzie efekt wstawienia kluczy 12, 44, 13, 88, 23, 94, 11, 39, 20, 16 i 5 do 11-elementowej tablicy asocjacyjnej,  
jeżeli funkcja mieszająca ma postać  
h(i) = (3i + 5) mod 11,  
a kolizje rozwiązywane są metodą łańcuchową?


```python
tablica = []
for i in range(11):
    tablica.append([])

def h(i):
    return (3*i+5)%11

klucze = [12,44,13,88,23,94,11,39,20,16,5]

for klucz in klucze:
    print("Klucz "+str(klucz)+ " będzie miał pozycję "+ str(h(klucz))+ " w tablicy asocjacyjnej.")
    tablica[h(klucz)].append(klucz)

print(tablica)
```

Zad. 2 Jaki będzie wynik poprzedniego zadania, jeżeli do rozwiązywania kolizji
użyjemy próbkowania liniowego?

```python
tablica = []
for i in range(11):
    tablica.append([])

def h(i):
    return (3*i+5)%11

klucze = [12,44,13,88,23,94,11,39,20,16,5]

for klucz in klucze:
    pozycja = h(klucz)
    while tablica[pozycja] != []:
        pozycja = (pozycja + 1)%11
    print("Klucz "+str(klucz)+ " będzie miał pozycję "+ str(pozycja)+ " w tablicy asocjacyjnej.")
    tablica[pozycja].append(klucz)

print(tablica)

```

Zad. 3 Jaki będzie wynik pierwszego zadania, jeżeli kolizje usuwane będą przy pomocy drugiej funkcji mieszającej  

h′(k) = 7 − (k mod 7)?


```python
tablica = []
for i in range(11):
    tablica.append([])

def h(i):
    return (3*i+5)%11

def h2(k):
    return 7 - (k%7)

klucze = [12,44,13,88,23,94,11,39,20,16,5]

for klucz in klucze:
    pozycja = h(klucz)
    if tablica[pozycja] != []:
        pozycja = h2(pozycja)
    print("Klucz "+str(klucz)+ " będzie miał pozycję "+ str(pozycja)+ " w tablicy asocjacyjnej.")
    tablica[pozycja].append(klucz)

print(tablica)
```

Zad. 6 Zaimplemetuj algorytm sortowania szybkiego w wersji bez rekurencji
oraz działającego w miejscu (ang. in-place), tzn. potrzebującego do wykonania stałego rozmiaru pamięci komputera, niezależnie od wielkości danych wejściowych.

```python
def quicksort(lista):
    stos = [(0, len(lista)-1)]
    
    while stos:
        low, high = stos.pop()
        if low < high:
            pivot_index = partition(lista,low,high)
            
            if pivot_index - low < high - pivot_index:
                stos.append((low,pivot_index-1))
                stos.append((pivot_index+1,high))
            else:
                stos.append((pivot_index+1,high))
                stos.append((low,pivot_index-1))
                
def partition(lista,low,high):
    pivot = lista[high]

    i = low-1

    for j in range(low,high):
        if lista[j] <= pivot:
            i += 1
            lista[i], lista[j] = lista[j], lista[i]
    lista[i+1], lista[high] = lista[high], lista[i+1]
    return i + 1

lista = [-1,24,2435,-123,545,2,43,22,-5,8]
print("oryginalna lista: ", lista)
quicksort(lista)
print("posortowana lista: ", lista)
```

```python
import random 
from matplotlib import pyplot as plt, animation, rc

  
  
# Sortowanie bąbelkowe
def bubblesort(A): 
    swapped = True
      
    for i in range(len(A) - 1): 
        if not swapped: 
            return
        swapped = False
          
        for j in range(len(A) - 1 - i): 
            if A[j] > A[j + 1]: 
                A[j], A[j+1] = A[j+1], A[j] 
                swapped = True
            yield A 
  
  
def visualize(): 
    N = 30
    A = list(range(1, N + 1)) 
    random.shuffle(A) 
      
    #generator, który zawiera wszystkie iteracje algorytmu sortującego
    generator = bubblesort(A) 
      
    # matplotlibowe rzeczy
    fig, ax = plt.subplots() 
    ax.set_title("Bubble Sort Animation") 
    bar_sub = ax.bar(range(len(A)), A, align="edge") 
    ax.set_xlim(0, N) 
    text = ax.text(0.02, 0.95, "", transform=ax.transAxes) 
    iteration = [0] 
      
    # funkcja pomocnicza żeby prostokąty się poprawnie wyświetlały i żeby numerator operacji działał jak natura chciała
    def update(A, rects, iteration): 
        for rect, val in zip(rects, A): 
            rect.set_height(val) 
        iteration[0] += 1
        text.set_text(f"# of operations: {iteration[0]}") 
  
    # tworzenie obiektu animacji
    anim = animation.FuncAnimation( 
        fig, 
        func=update, 
        fargs=(bar_sub, iteration), 
        frames=generator, 
        repeat=True, 
        blit=False, 
        interval=15, 
        save_count=90000, 
    ) 
    return anim

rc('animation', html='html5')
animacja = visualize()
animacja
```
