```python
def Pow1(x,n):
    """
    Funkcja wyliczająca x^n metodą naiwną (Pythonowe x^n ma złożoność O(nlogn), w zadaniu ma byc O(n^2))
    """
    pow = 1

    for i in range(n):
        pow = pow*x

    return pow

def Wielomian1(n,x0,a):
    """
    Funkcja wyliczająca wartość wielomianu w punkcie
    
    Input:
    n (int) - stopień wielomianu
    x0 (int) - punkt w którym obliczamy wartość wielomianu
    a (lista) - lista współczynników wielomianu, gdzie pozycja współczynnika na liście 
                odpowiada stopniu jednomianu, którego jest współczynnikiem.

    Output:
    sum (int) - wartość wielomianu w punkcie
    """
    sum = 0
    for i in range(n):
        sum += a[i]*(Pow1(x0,i))
    return sum
```

```python
def Wielomian2(n,x0,a):
    """
    Funkcja obliczająca wartość wielomianu w punkcie oparta na bardziej efektywnej metodzie obliczania x^i
    
    Input:
    n (int) - stopień wielomianu
    x0 (int) - punkt w którym obliczamy wartość wielomianu
    a (lista) - lista współczynników wielomianu, gdzie pozycja współczynnika na liście 
                odpowiada stopniu jednomianu, którego jest współczynnikiem.

    Output:
    sum (int) - wartość wielomianu w punkcie
    """
    sum = 0
    for i in range(n):
        sum += a[i]*(pow(x0,i))
    return sum
```

```python
def Horner(n, x0, a):
    """
    Funkcja obliczająca wartość wielomianu w punkcie oparta na metodzie Hornera
    
    Input:
    n (int) - stopień wielomianu
    x0 (int) - punkt w którym obliczamy wartość wielomianu
    a (lista) - lista współczynników wielomianu, gdzie pozycja współczynnika na liście 
                odpowiada stopniu jednomianu, którego jest współczynnikiem.

    Output:
    sum (int) - wartość wielomianu w punkcie
    """
    sum = a[n-1]
    for i in range(1,n):
        sum = x0*sum + a[n-(i+1)]
    return sum
```

```python
def example1(S):
    """Return the sum of the elements in sequence S."""
    n = len(S)
    total = 0
    for j in range(n):
        total += S[j]
    return total

def example2(S):
    """Return the sum of the elements with even index in sequence S.
    """
    n = len(S)
    total = 0
    for j in range(0, n, 2):
        total += S[j]
    return total

def example3(S):
    """Return the sum of the prex sums of sequence S."""
    n = len(S)
    total = 0
    for j in range(n):
        for k in range(1+j):
            total += S[k]
    return total

def example4(A, B): # assume that A and B have equal length
    """Return the number of elements in B equal to the sum of prex
    sums in A."""
    n = len(A)
    count = 0
    for i in range(n):
        total = 0
        for j in range(n):
            for k in range(1+j):
                total += A[k]
        if B[i] == total:
            count += 1
    return count
```

```python
import matplotlib.pyplot as plt
import random
import time
from scipy.optimize import curve_fit
import numpy as np
```

```python
test1 = random.sample(range(1,10000000), 100)
test2 = random.sample(range(1,10000000), 1000)
test3 = random.sample(range(1,10000000), 10000)
test4 = random.sample(range(1,10000000), 100000)
test5 = random.sample(range(1,10000000), 1000000)
```

```python
t1 = time.time()
example1(test1)
czas1 = time.time()-t1
t2 = time.time()
example1(test2)
czas2 = time.time()-t2
t3 = time.time()
example1(test3)
czas3 = time.time()-t3
t4 = time.time()
example1(test4)
czas4 = time.time()-t4
t5 = time.time()
example1(test5)
czas5 = time.time()-t5
```

```python
def f(x,a,b):
    return a*x+b

popt, pcov = curve_fit(f, [100,1000,10000,100000,1000000], [czas1,czas2,czas3,czas4,czas5])

x2 = np.arange(100,1000000)

plt.title("Example1")
plt.plot([100,1000,10000,100000,1000000],[czas1,czas2,czas3,czas4,czas5], 'ro', label="Dane")
plt.plot(x2, f(x2, *popt), label="Hipoteza")
plt.legend(loc='upper left')
plt.show()
```
